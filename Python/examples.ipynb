### Bragg Mirrors

```python
from Bragg import bragg
import numpy as np

# Parameters
lam = 1             # Central wavelength
n_High = 4          # High index
n_Low = 1.5         # Low index
n_inc = 1           # n of incident (typically air)
n_sub = 1.5         # n of substrate (ypically glass or same material)
m = 5               # Number of pairs
mesh_len = 1000     # Mesh length


## Different angles of incidence
theta_i0s = [0, np.deg2rad(30), np.deg2rad(60)] 

print("--- Varying angle of incidence ---")
for theta_i0 in theta_i0s:
    bragg(lam, n_High, n_Low, n_inc, n_sub, m, theta_i0, mesh_len)


print("--- Varying number of pairs ---")
## Different number of pairs
theta_i0 = 0
ms = [1, 2, 5, 10]

for m in ms:
    bragg(lam, n_High, n_Low, n_inc, n_sub, m, theta_i0, mesh_len)
```

### Fabry-Prrot resonant cavity


```python
from Fabry_perot import fabry_perot

## Parameters
lam0 = 1.0          
n_High = 3.5        
n_Low = 1.5         
n_cavity = 3.5      
n_inc = 1.0         
n_sub = 1.5                 
cav_mult = 2.0      
theta_i0 = 0.0      
mesh_len = 2000

ms = [1, 2, 4]
for m in ms:
    fabry_perot(lam0, n_High, n_Low, n_cavity, n_inc, n_sub, m, cav_mult, theta_i0, mesh_len)
```
